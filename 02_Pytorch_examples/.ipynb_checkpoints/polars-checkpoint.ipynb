```python
import polars as pl
from polars import selectors as cs
```

```python
df = pl.read_csv("used_cars.csv", separator=",")
```

```python
print(len(df))
```

```python
df = (
    df.with_columns(
        pl.col("price")
        .str.replace(r'\$', "")
        .str.replace(r',', '')
        .str.replace(r',', '')
        .cast(pl.Int64)   
    )
)
```

```python
df.filter(pl.col("price") < 300000)["price"].plot.hist()
#skewed towards normally price or used cars which makes sense obviously
```
