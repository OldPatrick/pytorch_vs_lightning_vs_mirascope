```python
import polars as pl
from polars import selectors as cs
```

```python
df = pl.read_csv("used_cars.csv", separator=",")
```

```python
df["model_year"].mean(), df["model_year"].min(), df["model_year"].median(), 
```

```python
df = (
    df.with_columns(
        pl.col("price")
        .str.replace(r'\$', "")
        .str.replace(r',', '')
        .str.replace(r',', '')
        .cast(pl.Int64)   
    )
)
```

```python
df
```

```python
df = (
    df
    .with_columns(
        pl.col('milage')
        .str.replace(r' mi.', '').str.replace(r',', '')
        .str.replace(r',', '')
        .cast(pl.Int64)  
        )
)
```

```python
df
```

```python
df.filter(pl.col("price") < 300000)["price"].plot.hist()
#skewed towards normally price or used cars which makes sense obviously
```

```python
df.filter(pl.col("brand").eq("Ford"))["price"].mean()
```

```python
df["brand"].to_dummies()
# in pandas it is get_dummies
#we could use the one hot encoded brand columns as predictors but we would ave quite an amount of columns for only 4000 data points
```

```python
df.select("model_year", "price").plot.scatter(x="model_year", y="price")
```
